- Приводим названия колонок к нижнему регистру.

- Переводим колонку "price" из типа float в тип int64. Предварительно проверив методом .unique() я определил, что в этой колонке нет знаков после точки, а значит и смысла в типе данных float нет.(Если бы знаки после запятой были, я бы все значения округлил и перевел в int)

# <a name="6">6. Проверка гипотез</a>

    H0: Средние пользовательские рейтинги красного и белого вина одинаковые.
    H1: Средние пользовательские рейтинги красного и белого вина разные.

```python
red = df.points[df.color=='red']
white = df.points[df.color=='white']
print('Средние пользовательские рейтинги красного вина =',red.mean().round(2),
      '\nСредние пользовательские рейтинги белого вина =',white.mean().round(2))
```

```python
levene, p = st.levene(red, white)
print('statistic=%.3f, pvalue=%.3f' % (levene, p))
alpha = 0.05
if p < alpha:
    print('Отклонить гипотезу о равенстве дисперсий')
else:
    print('Принять гипотезу о равенстве дисперсий')
```

```python
statistic, p = st.ttest_ind(red, white)
print('statistic=%.3f, p=%.3f' % (statistic, p))
alpha = 0.05
if p < alpha:
	print('Средние пользовательские рейтинги красного и белого вина статистически разные.')
else:
	print('Средние пользовательские рейтинги красного и белого вина статистически не различаются.')
```

    H0:Средние цены двух популярных сортов вина одинаковые.
    H1:Средние цены двух популярных сортов вина разные.

Как мы уже знаем, из подсчета рейтинга путем count, два самые популярных сорта вина это - Pinot Noir и Chardonnay. 
<br>
С ними я и буду работать

```python
a = df[df['variety'].isin(['Pinot Noir','Chardonnay'])]
Pinot_Noir = a.price[a.variety=='Pinot Noir']
Chardonnay = a.price[a.variety=='Chardonnay']
print('Средняя цена сорта Pinot Noir =',Pinot_Noir.mean().round(2),
      '\nСредняя цена сорта Chardonnay =',Chardonnay.mean().round(2))
```

```python
levene, p = st.levene(Pinot_Noir, Chardonnay)
print('statistic=%.3f, pvalue=%.3f' % (levene, p))
alpha = 0.05
if p < alpha:
    print('Отклонить гипотезу о равенстве дисперсий')
else:
    print('Принять гипотезу о равенстве дисперсий')
```

```python
statistic, p = st.ttest_ind(Pinot_Noir, Chardonnay)
print('statistic=%.3f, p=%.3f' % (statistic, p))
alpha = 0.05
if p < alpha:
	print('Средние цены двух популярных сортов вина статистически разные.')
else:
	print('Средние цены двух популярных сортов вина статистически не различаются.')
```
